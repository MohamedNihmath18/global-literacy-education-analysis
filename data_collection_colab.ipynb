```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries imported!")
```

```python
# Load all 5 datasets
df_adult = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-adults.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

df_youth = pd.read_csv("https://ourworldindata.org/grapher/literacy-rate-of-young-men-and-women.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

df_illit = pd.read_csv("https://ourworldindata.org/grapher/literate-and-illiterate-world-population.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

df_gdp = pd.read_csv("https://ourworldindata.org/grapher/gdp-per-capita-worldbank.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

df_school = pd.read_csv("https://ourworldindata.org/grapher/literacy-rates-vs-average-years-of-schooling.csv?v=1&csvType=full&useColumnShortNames=true",storage_options = {'User-Agent': 'Our World In Data data fetch/1.0'})

print("All 5 datasets loaded!")
print(f"Adult literacy  : {df_adult.shape}")
print(f"Youth literacy  : {df_youth.shape}")
print(f"Illiteracy pop  : {df_illit.shape}")
print(f"GDP per capita  : {df_gdp.shape}")
print(f"Years schooling : {df_school.shape}")
```

```python
# Preview all 5 datasets
for name, df in [("Adult Literacy",  df_adult),
                  ("Youth Literacy",  df_youth),
                  ("Illiteracy Pop",  df_illit),
                  ("GDP per Capita",  df_gdp),
                  ("Years Schooling", df_school)]:
    print(f"\n{'='*45}")
    print(f"{name}")
    print(f"Shape   : {df.shape}")
    print(f"Columns : {list(df.columns)}")
    print(df.head(3))
```

```python
# Quick data quality check
print("DATA QUALITY SUMMARY\n")
for name, df in [("Adult Literacy",  df_adult),
                  ("Youth Literacy",  df_youth),
                  ("Illiteracy Pop",  df_illit),
                  ("GDP per Capita",  df_gdp),
                  ("Years Schooling", df_school)]:
    print(f"{'─'*40}")
    print(f"{name}")
    print(f"   Rows         : {df.shape[0]}")
    print(f"   Columns      : {df.shape[1]}")
    print(f"   Missing vals : {df.isnull().sum().sum()}")
    print(f"   Year range   : {df.iloc[:,2].min()} – {df.iloc[:,2].max()}")
    print(f"   Countries    : {df.iloc[:,0].nunique()}")
```

```python
# Save raw datasets as CSV files
df_adult.to_csv('adult_literacy_raw.csv',   index=False)
df_youth.to_csv('youth_literacy_raw.csv',   index=False)
df_illit.to_csv('illiteracy_pop_raw.csv',   index=False)
df_gdp.to_csv('gdp_per_capita_raw.csv',     index=False)
df_school.to_csv('years_schooling_raw.csv', index=False)

print("All 5 raw CSV files saved in Colab!")
```
